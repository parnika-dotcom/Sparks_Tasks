##**Ananlysis of sales and Profit with respect to different States**


```
state=df.groupby("State").sum()
state_sale=state.sort_values(by=("Sales"),ascending=False)
plt.figure(figsize=(20,20))
plt.xticks(size=10)
plt.yticks(size=10)
sns.barplot(x="Sales", y=state_sale.index,data=state_sale)
plt.show()
```

**We see the sales are very high for the states California, New York and Texas. However the sales are very low for states like Maine, West Virginia, North datoka.**

```
state_profit=state.sort_values(by=("Profit"),ascending=False)
plt.figure(figsize=(20,20))
plt.xticks(size=10)
plt.yticks(size=10)
sns.barplot(x="Profit", y=state_profit.index,data=state_profit)
plt.show()
```

**The states California, New York, washington are the states giving high profit to the business.**

**However the states Texas, Ohio ,Pennsylvania, Illinois, North Carolina, Colorado, Tennessee, Arizona, Florida , Oregon,Wyoming, West Virginia provide negative profit to the business.**

**Although the states Texas, Ohio ,Pennsylvania, Illinois, North Carolina have quite good percentage of sale.**
