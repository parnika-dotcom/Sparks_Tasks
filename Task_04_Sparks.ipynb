##**Cleaning the Data**

```
#changing the name of the old columns to new ones.
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```
#we only take important columns which are required only in our dataframe.
df1=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```
df1.head()
```

##**Data Visualization**

```
#Check the corelation between the different parameters.
sns.pairplot(df1, hue = 'Region')
plt.show()
```

```
#Count of the number of Countries that are affected by Terrorist attacks are representation in the form of top five and last less attacks by terrorist
df['Country'].value_counts().head()
```

```
df['Country'].value_counts().tail()
```

```
df['Country'].value_counts()
plt.figure(figsize=(10,8))
sns.barplot(x=df['Country'].value_counts()[:20].index,y=df['Country'].value_counts()[:20].values)
plt.title('Top Countries got attacked')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('No. of Attacks')
plt.show()
```

**Here, we can conclude that the number of Attacks V/s the Countries representation of the graph that EL Salvador has high number of attacks and least in South Africa country.**

##**Visualization representation of the region were mostly affected by Terrorist attacks.**

```
print('Region were the most attacks are found: '+df['Region'].value_counts().idxmax())
```

```
df['Region'].value_counts()
plt.figure(figsize=(10,8))
sns.countplot(x='Region',data=df)
plt.xticks(rotation=90)
plt.show()
```

###**Here, we can conclude from the Count V/s Region representation of the graph that the Region - Central America & Caribbean are of higher attacks and least attacks in the South Africa country.**

**The Cities were more affected by terrorist attacks.**

```
df['city'].value_counts()
```

```
df['city'].value_counts()
plt.figure(figsize=(10,8))
sns.barplot(x=df['city'].value_counts()[:20].index,y=df['city'].value_counts()[:20].values)
plt.title('Top Cities got attacks')
plt.xticks(rotation=90)
plt.xlabel('Cities')
plt.ylabel('No of Attacks')
plt.show()
```

###**Here,we conclude that No of Attacks V/s Cities as there are top 20 cities which got highest attacks.**

**Visualization based on the Trend of attacks form 1970 to 2017 Years with the count representation.**


```
plt.subplots(figsize=(10,8))
sns.countplot(x='Year',data=df,palette='tab10')
plt.xticks(rotation=90)
plt.title('Terror attacks from 1970 TO 2017')
plt.show()
```

###**Here,we conclude that the terror attacks from 1979 TO 1982 had an immense increase in the number of  attacks.**

**Visualization representation of the type of attack used by Terrorists**

```
plt.figure(figsize=(10,8))
sns.countplot(x=df['AttackType'],order=df['AttackType'].value_counts().index,palette='rainbow')
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()
```

###**Here,we conclude that the Number of Attacks V/S Attack type as mostly using by Bombing/Explosion are formed and least type of attack are used by Unarmed Assault.**

**Visualization based on the Weapons which are used for supporting for the type of attack.**

```
df['Weapon_type'].value_counts()
plt.figure(figsize=(10,8))
color=['blue','red','green','orange']
sns.barplot(x=df['Weapon_type'].value_counts()[:20].index,y=df['Weapon_type'].value_counts()[:20].values , palette=color)
plt.title('Weapons Usage')
plt.xticks(rotation=90)
plt.xlabel('Weapon used')
plt.ylabel('No fo times')
plt.show()
```

```
#Weapon used for supporting the attack
df['Target_type'].value_counts()
```

```
#Visualization based on the number of the Attacks targets and Targets Categories
plt.figure(figsize=(10,8))
sns.countplot(x=df['Target_type'],order=df['Target_type'].value_counts().index,palette='rainbow')
plt.xticks(rotation=90)
plt.xlabel('Targets')
plt.ylabel('No of Targets')
plt.show()
```

###**Here,we can conclude that No of Targets V/s the Targets are mostly by using Terrorist target towards the Business person and least by Abortion Related other etc.**

**Visualization on the bases of the Name of the group attacks and No.of attacks by Terrorist.**

```
plt.figure(figsize=(15,6))
sns.barplot(x=df['Group'].value_counts()[:20].index,y=df['Group'].value_counts()[:20].values,palette='pastel')
plt.xticks(rotation=90)
plt.title('Total Group Attacks')
plt.xlabel('Group Name')
plt.ylabel('Total no. of attacks')
plt.show()
```

**Here,we conclude that the Total No.of attacks V/s the Group Name as there are Unknown group attack mostly and least by the National Liberation Armay of Colombia(ELN).**
